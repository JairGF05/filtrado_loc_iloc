# **Uso de Loc**
puedo filtrar por algunas filas solo de esos atributos que estoy seleccionando, usando Loc que filtra segun label.

```python
df_books.loc[:]
```

```python
#Aplicando Loc del 0 al 4
df_books.loc[0:4]
```

```python
#Ahora puedo especificar que me muestre del 0 al 4 pero únicamente por nombre y autor
df_books.loc[0:4, ['Name', 'Author']]
```

```python
#Fltrar unicamente todas las reviews y multiplicarlas por -1
df_books.loc[:,['Reviews']] *-1
```

```python
#Filtrar por autor cuando sea igual a JJ Smith
df_books.loc[:, ['Author']] == 'JJ Smith'
```

# **Usando Iloc**
Filtra mediante Índices

```python
df_books.iloc[:]
```

```python
#Ahora especifico que quiero todas las filas, pero de la columna 0 a 3
df_books.iloc[:, 0:3]
```

```python
#Modificar un valor en especifico
#De la row 1 , quiero modificar el valor 3, en este caso volverlo negativo
df_books.iloc[1, 3] * -1
```

```python
#Puedo seleccionar solo las primeras 2 filas y 2 columnas
df_books.iloc[:2, 2:]
```
